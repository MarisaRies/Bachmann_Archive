```python
# import required packages
import os
import tellurium as te
import phrasedml
```

```python
# set working directory - change file path for .ipynb manually as required
workingDir = '/content/drive/My Drive/bachmann_archiv/Group4_GenerationSEDML/Tellurium/experiment/'
os.chdir(workingDir)
```

```python
# retrieve SBML model from model folder and set as reference for phraSEDML
sbml = "../model/Bachmann2011.xml"
ant_str = te.sbmlToAntimony(sbml)
sbml_str = te.antimonyToSBML(ant_str)
phrasedml.setReferencedSBML('../model/Bachmann2011.xml', sbml_str)
# print(sbml_str)
```

```python
# create phraSEDML string as template for SED-ML
phrasedml_str = """
    model1 = model "{}"
    sim1 = simulate uniform(0, 250, 1000)
    task1 = run sim1 on model1

    plot "Supplementary Figure 9 EpoRJAK2" task1.time vs task1.EpoRJAK2
    plot "Supplementary Figure 9 EpoRpJAK2" task1.time vs task1.EpoRpJAK2
    plot "Supplementary Figure 9 p1EpoRpJAK2" task1.time vs task1.p1EpoRpJAK2
    plot "Supplementary Figure 9 p2EpoRpJAK2"task1.time vs task1.p2EpoRpJAK2
    plot "Supplementary Figure 9 p12EpoRpJAK2"task1.time vs task1.p12EpoRpJAK2
    plot "Supplementary Figure 9 SHP1" task1.time vs task1.SHP1
    plot "Supplementary Figure 9 SHP1Act" task1.time vs task1.SHP1Act
    plot "Supplementary Figure 9 STAT5" task1.time vs task1.STAT5
    plot "Supplementary Figure 9 pSTAT5" task1.time vs task1.pSTAT5
    plot "Supplementary Figure 9 npSTAT5" task1.time vs task1.npSTAT5
    plot "Supplementary Figure 9 CISRNA" task1.time vs task1.CISRNA
    plot "Supplementary Figure 9 CIS" task1.time vs task1.CIS
    plot "Supplementary Figure 9 SOCS3RNA" task1.time vs task1.SOCS3RNA
    plot "Supplementary Figure 9 SOCS3" task1.time vs task1.SOCS3
    

""".format('../model/Bachmann2011.xml')
```

```python
# convert to SED-ML and create SED-ML file in working directory
sedml_str = phrasedml.convertString(phrasedml_str)
if sedml_str == None:
    raise RuntimeError(phrasedml.getLastError())
print(sedml_str)

with open(os.path.join(workingDir, 'SuppFig9.sedml'), 'wb') as f:
    f.write(sedml_str.encode('utf-8'))
```

```python
# run the SED-ML files based on the referenced SBML file in ../models/ 
te.executeSEDML(sedml_str)
```
